```python
ground_truth = 10
x = 3
w = 1.0
b = 0.0
learning_rate = 0.05

for i in range(10):
  prediction = w * x + b
  loss = (prediction - ground_truth)**2
  print(f"Step {i+1}: prediction={prediction:.2f}, loss={loss:.2f}, w={w:.3f}, b={b:.3f}")
  w += learning_rate * (ground_truth - prediction) * x
  b += learning_rate * (ground_truth - prediction)
```
